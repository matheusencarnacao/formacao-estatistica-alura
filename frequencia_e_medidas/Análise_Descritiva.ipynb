### Para avaliarmos o comportamento da variável RENDA vamos construir uma tabela de frequências considerando as seguintes classes em salários mínimos (SM)
#### <font color='blue'>Descreva os pontos mais relevantes que você observa na tabela e no gráfico.</font>

Classes de renda:

<b>A</b> ► Acima de 25 SM

<b>B</b> ► De 15 a 25 SM

<b>C</b> ► De 5 a 15 SM

<b>D</b> ► De 2 a 5 SM

<b>E</b> ► Até 2 SM

Para construir as classes de renda considere que o salário mínimo na época da pesquisa era de <b>R$ 788,00</b>.

#### Siga os passos abaixo:

### 1º Definir os intevalos das classes em reais (R$)

```python
intervalos = [0, 1576, 3940, 11820, 19700, 200000]
```

### 2º Definir os labels das classes

```python
labels = ['E', 'D', 'C', 'B', 'A']
```

### 3º Construir a coluna de frequências

```python
frequencia = pd.value_counts(pd.cut(x = df['Renda'], bins=intervalos, labels=labels, include_lowest=True))
frequencia
```

### 4º Construir a coluna de percentuais

```python
percentual = pd.value_counts(pd.cut(x = df['Renda'], bins=intervalos, labels=labels, include_lowest=True),  normalize=True)
percentual
```

### 5º Juntar as colunas de frequência e percentuais e ordenar as linhas de acordo com os labels das classes

```python
df_freq_percent = pd.DataFrame(
    {'Frequência': frequencia, 'Porcentagem (%)': percentual}
)
df_freq_percent
```

### Construa um gráfico de barras para visualizar as informações da tabela de frequências acima

```python
df_freq_percent.plot.bar(width= 1, alpha = 0.2, figsize= (12, 6))
```

> ### Conclusões

Escreva suas conclusões aqui...

60% da população ganha menos que 2 salarios minimos e a quantidade de pessoas que ganham mais de 25 salarios minimos, é praticamente 0% (0,005%)

### Obtenha a média, mediana e valor máximo da variável RENDA segundo SEXO e COR
#### <font color='yellow'>Destaque os pontos mais importante que você observa nas tabulações</font>
#### <font color='red'>O parâmento <i>aggfunc</i> da função <i>crosstab()</i> pode receber uma lista de funções. Exemplo: <i>aggfunc = {'mean', 'median', 'max'}</i></font>

```python
agg_sexo_cor = pd.crosstab(index=df.Cor, columns=df.Sexo, values=df.Renda, aggfunc={'mean', 'median', 'max'})
agg_sexo_cor.rename(index=cor, inplace=True)
agg_sexo_cor.rename(columns=sexo, inplace=True)
agg_sexo_cor
```

> ### Conclusões

Escreva suas conclusões aqui...

A maioria dos homens de Cor amarela ganham mais que as outras Cores, e a maioria das mulheres amarelas ganham mais que as de outras cores. Porem os maiores salarios sao de homens e mulheres brancas.

### Obtenha as medidas de dispersão da variável RENDA segundo SEXO e COR
#### <font color='yellow'>Destaque os pontos mais importante que você observa nas tabulações</font>
#### <font color='red'>O parâmento <i>aggfunc</i> da função <i>crosstab()</i> pode receber uma lista de funções. Exemplo: <i>aggfunc = {'mad', 'var', 'std'}</i></font>

```python
agg_sexo_cor = pd.crosstab(index=df.Cor, columns=df.Sexo, values=df.Renda, aggfunc={'mad', 'var', 'std'})
agg_sexo_cor.rename(index=cor, inplace=True)
agg_sexo_cor.rename(columns=sexo, inplace=True)
agg_sexo_cor
```

> ### Conclusões

Escreva suas conclusões aqui...

Ha uma discrepancia muito grande na renda de mulheres indigenas, e por isso vemos que ha um desvio padrao muito mais alto que de outras categorias.

Tambem analisamos que os homens costumam ter um  desvio padrao maior que dos das mulheres da mesma categoria.
